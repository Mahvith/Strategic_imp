```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.calibration import CalibratedClassifierCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# 1. Load Data
X_train = pd.read_csv('adult_Xtrain.csv')
y_train = pd.read_csv('adult_ytrain.csv')
X_test = pd.read_csv('adult_Xtest.csv')
y_test = pd.read_csv('adult_ytest.csv')

# 2. Setup Pipeline
# Numerical features to scale
num_features = ['age', 'fnlwgt', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week']
# Categorical features to encode
cat_features = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'gender', 'native-country']

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_features),
        ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), cat_features)
    ])

# 3. Train Calibrated Model
lr_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('clf', CalibratedClassifierCV(
        estimator=LogisticRegression(max_iter=2000, random_state=42),
        method='sigmoid', cv=5
    ))
])

lr_pipeline.fit(X_train, y_train.values.ravel())

# 4. Visualization Function
def plot_monotonicity(model, X_data, feature_names):
    n = len(feature_names)
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    axes = axes.flatten()

    # Use a subset for cleaner ICE plots
    X_sample = X_data.sample(n=50, random_state=42).copy()

    for i, feature in enumerate(feature_names):
        ax = axes[i]
        min_v, max_v = X_data[feature].min(), X_data[feature].max()

        # Grid of values to test
        if feature == 'education-num':
            grid = np.sort(X_data[feature].unique())
        else:
            grid = np.linspace(min_v, max_v, 20)

        # Generate Predictions (ICE)
        ice_preds = []
        for val in grid:
            X_temp = X_sample.copy()
            X_temp[feature] = val
            ice_preds.append(model.predict_proba(X_temp)[:, 1])

        ice_preds = np.array(ice_preds).T
        avg_pred = np.mean(ice_preds, axis=0)

        # Plot
        ax.plot(grid, ice_preds.T, color='blue', alpha=0.05) # Individual lines
        ax.plot(grid, avg_pred, color='red', linewidth=3, label='Avg Effect') # Average
        ax.set_title(f"Monotonicity: {feature}")
        ax.set_xlabel(feature)
        ax.set_ylabel("P(Income > 50K)")
        ax.legend()
        ax.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

# 5. Run
features = ["education-num", "capital-gain", "age", "hours-per-week"]
plot_monotonicity(lr_pipeline, X_test, features)
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import log_loss

# 1. Load Data
X_train = pd.read_csv('adult_Xtrain.csv')
y_train = pd.read_csv('adult_ytrain.csv')
X_test = pd.read_csv('adult_Xtest.csv')
y_test = pd.read_csv('adult_ytest.csv')

# 2. Select strictly monotonic features
features = ["education-num", "capital-gain", "age", "hours-per-week"]
X_train_sel = X_train[features]
X_test_sel = X_test[features]

# 3. Define Pipeline
# StandardScaler is CRITICAL for linear models to find optimal weights
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('clf', LogisticRegression(solver='lbfgs', max_iter=2000, random_state=42))
])

# 4. Search for Optimal Regularization (C)
# 'C' controls regularization strength. High C = Low regularization (closer to pure MLE).
param_grid = {'clf__C': [0.1, 1, 10, 100, 1000]}

grid = GridSearchCV(
    pipeline,
    param_grid,
    scoring='neg_log_loss',
    cv=5
)

print("Optimizing for Minimum Log Loss...")
grid.fit(X_train_sel, y_train.values.ravel())

# 5. Get Results
best_model = grid.best_estimator_
probs = best_model.predict_proba(X_test_sel)[:, 1]
loss = log_loss(y_test, probs)

# Extract Weights (here function is of form: f(x) = w^T x + b)
weights = best_model.named_steps['clf'].coef_[0]
intercept = best_model.named_steps['clf'].intercept_[0]

print(f"\nOptimization Complete.")
print(f"Optimal Test Log Loss: {loss:.5f}")
print("-" * 30)
print(f"Intercept: {intercept:.4f}")
for feat, w in zip(features, weights):
    print(f"Weight ({feat}): {w:.4f}")
```

## Learning optimal linear classifier

```python
import torch
import torch.nn as nn
from torch.utils.data import TensorDataset, DataLoader
from torch.optim import Adam
from torch.optim.lr_scheduler import ReduceLROnPlateau
import numpy as np
from tqdm import tqdm
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# --- 1. Device Setup ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Training on: {device}")

# --- 2. Model Setup ---
class LinearClassifier(nn.Module):
    def __init__(self, in_dim, out_dim):
        super().__init__()
        self.out = nn.Linear(in_dim, out_dim, bias=True)

    def forward(self, x):
        return self.out(x)

# --- 3. Trainer Class with Validation ---
class Trainer:
    def __init__(self, model, train_dl, val_dl, opt, scheduler):
        self.model = model
        self.train_dl = train_dl
        self.val_dl = val_dl
        self.opt = opt
        self.scheduler = scheduler

    def loss(self, xb, y_tilde):
        """
        Naive Hinge loss
        """
        # Model prediction (B,)
        pred = self.model(xb).squeeze(-1)
        margin = y_tilde * pred
        loss = torch.relu(1 - margin).mean()
        return loss

    def train(self, n_epochs: int):
        best_val_loss = float("inf")
        history = {'train_loss': [], 'val_loss': []}

        for epoch in tqdm(range(n_epochs)):
            # --- Training Phase ---
            self.model.train()
            train_batch_losses = []

            for xb, yb in self.train_dl:
                self.opt.zero_grad(set_to_none=True)

                xb = xb.to(device)
                yb = yb.to(device).view(-1).long()
                y_tilde = (2 * yb - 1).float()
                loss = self.loss(xb, y_tilde)

                loss.backward()
                self.opt.step()
                train_batch_losses.append(loss.item())

            avg_train_loss = np.mean(train_batch_losses)
            history['train_loss'].append(avg_train_loss)

            # --- Validation Phase ---
            self.model.eval()
            val_batch_losses = []

            with torch.no_grad():
                for xb, yb in self.val_dl:
                    xb = xb.to(device)
                    yb = yb.to(device).view(-1).long()
                    y_tilde = (2 * yb - 1).float()
                    val_loss = self.loss(xb, y_tilde)
                    val_batch_losses.append(val_loss.item())

            avg_val_loss = np.mean(val_batch_losses)
            history['val_loss'].append(avg_val_loss)

            # # --- Scheduler & Checkpointing ---
            self.scheduler.step(avg_val_loss)

            if avg_val_loss < best_val_loss:
                best_val_loss = avg_val_loss
                torch.save(self.model.state_dict(), 'best_model_naive.pth')

            # --- Logging ---
            if epoch % 20 == 0:
                # Multi-D logging: Print norm of weights instead of single threshold
                w_norm = torch.norm(self.model.out.weight).item()
                b_val = self.model.out.bias.item()
                print(f"Ep: {epoch:04d} | Train: {avg_train_loss:.4f} | Val: {avg_val_loss:.4f} | |w|: {w_norm:.4f} | b: {b_val:.4f}")

        return history
```

```python
# --- Setup & Execution ---

# 1. Load Training Data
X_train = pd.read_csv('adult_Xtrain.csv')
y_train = pd.read_csv('adult_ytrain.csv')

# 2. Select strictly monotonic features
features = ["education-num", "capital-gain", "age", "hours-per-week"]
X_train_sel = X_train[features]

# 1. Preprocessing (CRITICAL)
# You MUST standardize your inputs before passing them to the model
# using the mean/std from your original training set.
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_sel) # Use this for training!

# We want shape (N, 4)
# X_train_scaled is ALREADY a numpy array.
X = X_train_scaled.astype(np.float32)            # Shape: (21113, 4)
Y = y_train.values.astype(np.float32) # y_train is likely still a DataFrame/Series  # Shape: (21113, 1)

# Ensure Y is (N, 1) if it isn't already
if Y.ndim == 1:
    Y = Y.reshape(-1, 1)

print(f"X Shape: {X.shape}") # Should be (21113, 4)
print(f"Y Shape: {Y.shape}") # Should be (21113, 1)

# Split 90/10
X_train_np, X_val_np, Y_train_np, Y_val_np = train_test_split(X, Y, test_size=0.2,
                                                              random_state=42, stratify=Y)

# Convert to Tensors
X_train_t = torch.from_numpy(X_train_np)
Y_train_t = torch.from_numpy(Y_train_np)
X_val_t = torch.from_numpy(X_val_np)
Y_val_t = torch.from_numpy(Y_val_np)

# 2. Optimized DataLoaders
train_ds = TensorDataset(X_train_t, Y_train_t)
val_ds = TensorDataset(X_val_t, Y_val_t)

# Define pin_memory based on device availability
use_pin_memory = torch.cuda.is_available()

train_dl = DataLoader(train_ds, batch_size=128, shuffle=True, pin_memory=use_pin_memory)
val_dl = DataLoader(val_ds, batch_size=128, shuffle=False, pin_memory=use_pin_memory)

print("Data successfully loaded!")

# 3. Model Setup
model_naive = LinearClassifier(in_dim=4, out_dim=1).to(device)

# Compilation (Optional)
try:
    model_naive = torch.compile(model_naive)
except:
    pass

# SGD Optimizer
opt = torch.optim.SGD(model_naive.parameters(), lr=0.01, momentum=0.9, weight_decay=0.0)

# (StepLR is great for convergence)
# Decays learning rate by 0.1 every time validation loss plateaus
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
    opt, mode='min', factor=0.5, patience=20
) ## if want, can change here later, only!

# 5. Initialize and Train
trainer_naive = Trainer(
    model=model_naive,
    train_dl=train_dl,
    val_dl=val_dl,
    opt=opt,
    scheduler=scheduler
)

history = trainer_naive.train(100)
```

## Learning Strategic Classifier

```python
import torch
import torch.nn as nn
from torch.utils.data import TensorDataset, DataLoader
from torch.optim import Adam
from torch.optim.lr_scheduler import ReduceLROnPlateau
import numpy as np
from tqdm import tqdm
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
@torch.no_grad()
def find_best_response_batch(xb, samples, model, beta, alpha):
    '''
    Finds the best strategic move for a batch of agents xb against a set of samples.
    Adaptable for D-dimensional data.

    Args:
        xb: (B, D) Batch of agents
        samples: (N, D) Database of potential moves
        model: maps (N, D) -> (N,) scores
        beta: scalar budget constraint
        alpha: (D,) vector of cost weights per feature
    '''
    device = xb.device
    dtype = xb.dtype

    # 1. Handle Dimensions (Ensure 2D shapes)
    # If input is 1D (B,), make it (B, 1)
    if xb.dim() == 1:
        xb = xb.unsqueeze(dim=1)
    if samples.dim() == 1:
        samples = samples.unsqueeze(dim=1)

    B, D = xb.shape
    N, D_sample = samples.shape

    # Sanity Check
    assert D == D_sample, f"Dimension Mismatch: Agents have {D} features, Samples have {D_sample}"

    # 2. Process Alpha (Cost Weights)
    # We need alpha to be shape (1, D) to broadcast against (B, N, D)
    alpha_t = torch.as_tensor(alpha, device=device, dtype=dtype)

    if alpha_t.numel() == 1:
        # If scalar alpha given for multi-D, broadcast it to all dims
        alpha_t = alpha_t.view(1, 1).expand(1, D)
    else:
        assert alpha_t.numel() == D, f"Alpha vector must have length {D}, got {alpha_t.numel()}"
        alpha_t = alpha_t.view(1, D)

    beta_t = torch.as_tensor(beta, device=device, dtype=dtype)

    # 3. Calculate Scores for all Samples (Potential Destinations)
    # This gives us the 'utility' of every point in the database
    sample_scores = model(samples).view(-1) # shape: (N,)

    # 4. Calculate Cost Matrix (Vectorized for Multi-D)
    # diff shape: (B, N, D)
    # Broadcasting: (B, 1, D) vs (1, N, D) -> (B, N, D)
    diff = samples.unsqueeze(0) - xb.unsqueeze(1)

    # Cost Function: sum_i( alpha_i * max(0, z_i - x_i) )
    # .clamp(min=0) implements the max(0, ...)
    # .sum(dim=2) aggregates costs across all D features
    cost_matrix = (diff.clamp(min=0) * alpha_t).sum(dim=2) # Result shape: (B, N)

    # 5. Mask Feasible Moves
    is_positive = sample_scores.unsqueeze(0) > 0 # (1, N)
    is_within_budget = cost_matrix <= beta_t     # (B, N)
    feasible_mask = is_positive & is_within_budget

    # 6. Select Best Strategic Move
    # We clone scores to create a masked version for each agent in batch
    masked_scores = sample_scores.unsqueeze(0).expand(B, N).clone()

    # Set infeasible moves to -infinity so they are never chosen
    masked_scores[~feasible_mask] = -float('inf')

    # Find the index of the max score for each agent
    best_indices = masked_scores.argmax(dim=1) # (B,)
    best_strategies = samples[best_indices]    # (B, D)

    # Check if a valid move was actually found (score != -inf)
    max_scores = masked_scores.gather(1, best_indices.unsqueeze(1)).squeeze(1)
    found_feasible = torch.isfinite(max_scores) # (B,)

    # 7. Final Selection
    # If feasible found -> move there. Else -> stay at original position (xb)
    # Unsqueeze found_feasible to (B, 1) to broadcast across D features
    final_strategies = torch.where(found_feasible.unsqueeze(1), best_strategies, xb)

    return final_strategies
```

```python
# --- 1. Device Setup ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Training on: {device}")

# --- 2. Model Setup ---
class LinearClassifier(nn.Module):
    def __init__(self, in_dim, out_dim):
        super().__init__()
        self.out = nn.Linear(in_dim, out_dim, bias=True)

    def forward(self, x):
        return self.out(x)

# --- 3. Trainer Class with Validation ---
class StrategicTrainer:
    def __init__(self, model, train_dl, val_dl, opt, scheduler, alpha, beta, samples):
        self.model = model
        self.train_dl = train_dl
        self.val_dl = val_dl
        self.opt = opt
        self.scheduler = scheduler

        # Handle Alpha (Cost Weights) for Multi-D
        # Ensure alpha is a tensor of shape (D,) on the correct device
        if not torch.is_tensor(alpha):
            alpha = torch.tensor(alpha, dtype=torch.float32)
        self.alpha = alpha.to(device)

        self.beta = beta
        self.samples = samples.to(device)

    def strat_loss(self, xb, y_tilde):
        """
        Strategic Improvement-Aware Hinge Loss (Multi-D).
        """
        # Model prediction (B,)
        pred = self.model(xb).squeeze(-1)
        w = self.model.out.weight.squeeze() # Shape (D,)

        # Strategic Gain (Scalar)
        # In Multi-D linear model, agent puts budget into feature with max (w_i / alpha_i)
        # We clamp w to 0 because agents won't increase a feature if it has negative weight
        # (Assuming cost is for increasing feature value)
        'can have without torch.relu() the weights,also'
        roi = torch.relu(w) / (self.alpha + 1e-8)
        # roi = w / (self.alpha + 1e-8)
        max_roi = torch.max(roi)
        s_gain = self.beta * max_roi

        # Hinge-like loss with strategic gap
        # If y_tilde is 1 (Positive/Improved), we want Score > 1 - s_gain
        # "Give credit for the potential gain"
        margin = y_tilde * (pred + s_gain)
        loss = torch.relu(1 - margin).mean()
        return loss

    def train(self, n_epochs: int):
        best_val_loss = float("inf")
        history = {'train_loss': [], 'val_loss': []}

        for epoch in tqdm(range(n_epochs)):
            # --- Training Phase ---
            self.model.train()
            train_batch_losses = []

            for xb, yb in self.train_dl:
                self.opt.zero_grad(set_to_none=True)

                xb = xb.to(device)
                yb = yb.to(device).view(-1).long()
                y_tilde = (2 * yb - 1).float()
                loss = self.strat_loss(xb, y_tilde)

                loss.backward()
                self.opt.step()
                train_batch_losses.append(loss.item())

            avg_train_loss = np.mean(train_batch_losses)
            history['train_loss'].append(avg_train_loss)

            # --- Validation Phase ---
            self.model.eval()
            val_batch_losses = []

            with torch.no_grad():
                for xb, yb in self.val_dl:
                    xb = xb.to(device)
                    yb = yb.to(device).view(-1).long()
                    y_tilde = (2 * yb - 1).float()
                    val_loss = self.strat_loss(xb, y_tilde)
                    val_batch_losses.append(val_loss.item())

            avg_val_loss = np.mean(val_batch_losses)
            history['val_loss'].append(avg_val_loss)

            # # --- Scheduler & Checkpointing ---
            # self.scheduler.step(avg_val_loss)

            if avg_val_loss < best_val_loss:
                best_val_loss = avg_val_loss
                torch.save(self.model.state_dict(), 'best_model_strategic_adult.pth')

            # --- Logging ---
            if epoch % 10 == 0:
                # Multi-D logging: Print norm of weights instead of single threshold
                w_norm = torch.norm(self.model.out.weight).item()
                b_val = self.model.out.bias.item()
                print(f"Ep: {epoch:04d} | Train: {avg_train_loss:.4f} | Val: {avg_val_loss:.4f} | |w|: {w_norm:.4f} | b: {b_val:.4f}")

        return history
```

```python
# --- Setup & Execution ---

# 1. Load Training Data
X_train = pd.read_csv('adult_Xtrain.csv')
y_train = pd.read_csv('adult_ytrain.csv')

# 2. Select strictly monotonic features
features = ["education-num", "capital-gain", "age", "hours-per-week"]
X_train_sel = X_train[features]

# 1. Preprocessing (CRITICAL)
# You MUST standardize your inputs before passing them to the model
# using the mean/std from your original training set.
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_sel) # Use this for training!

# We want shape (N, 4)
# X_train_scaled is ALREADY a numpy array.
X = X_train_scaled.astype(np.float32)            # Shape: (21113, 4)
Y = y_train.values.astype(np.float32)          # Shape: (21113, 1)

# Ensure Y is (N, 1) if it isn't already
if Y.ndim == 1:
    Y = Y.reshape(-1, 1)

print(f"X Shape: {X.shape}") # Should be (21113, 4)
print(f"Y Shape: {Y.shape}") # Should be (21113, 1)

# Split 80/20
X_train_np, X_val_np, Y_train_np, Y_val_np = train_test_split(X, Y, test_size=0.2,
                                                              random_state=42, stratify=Y)

# Convert to Tensors
X_train_t = torch.from_numpy(X_train_np)
Y_train_t = torch.from_numpy(Y_train_np)
X_val_t = torch.from_numpy(X_val_np)
Y_val_t = torch.from_numpy(Y_val_np)

# We also need the full X tensor for the 'samples' argument if agents can move anywhere in distribution
# X_full_t = torch.from_numpy(X).to(device)
# # Note: passing X_full_t as 'samples' so agents can reference the full distribution if needed (data leakage issue )

# 2. Optimized DataLoaders
train_ds = TensorDataset(X_train_t, Y_train_t)
val_ds = TensorDataset(X_val_t, Y_val_t)

# Define pin_memory based on device availability
use_pin_memory = torch.cuda.is_available()

train_dl = DataLoader(train_ds, batch_size=32, shuffle=True, pin_memory=use_pin_memory)
val_dl = DataLoader(val_ds, batch_size=32, shuffle=False, pin_memory=use_pin_memory)

print("Data successfully loaded!")

# 3. Model Setup
mlp_oned_strat = LinearClassifier(in_dim=4, out_dim=1).to(device)

# Compilation (Optional)
try:
    mlp_oned_strat = torch.compile(mlp_oned_strat)
except:
    pass

# # 4. AdamW & Scheduler
opt = Adam(mlp_oned_strat.parameters(), lr = 0.001)
scheduler = ReduceLROnPlateau(opt, mode='min', factor=0.5, patience=20)

# SGD Optimizer
# opt = torch.optim.SGD(mlp_oned_strat.parameters(), lr=0.01, momentum=0.9, weight_decay=1e-4)

# # (StepLR is great for convergence)
# # Decays learning rate by 0.1 every time validation loss plateaus
# scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
#     opt, mode='min', factor=0.1, patience=5
# )

# 5. Initialize and Train
# The agents in the training loop can only "see" or "move to" other training examples, so samples = X_train_t
trainer_imp_aware = StrategicTrainer(
    model=mlp_oned_strat,
    train_dl=train_dl,
    val_dl=val_dl,
    opt=opt,
    scheduler=scheduler,
    alpha=np.array((1, 1, 1,1)),
    beta=1.0,
    samples=X_train_t
)

history = trainer_imp_aware.train(100)
```

## Using Improvement-Aware Algorithm

```python
import torch
import torch.nn as nn
from torch.utils.data import TensorDataset, DataLoader
from torch.optim import Adam
from torch.optim.lr_scheduler import ReduceLROnPlateau
import numpy as np
from tqdm import tqdm
from sklearn.model_selection import train_test_split
import warnings
from sklearn.exceptions import InconsistentVersionWarning

# Suppress the specific sklearn feature name warning
warnings.filterwarnings("ignore", category=UserWarning, module="sklearn")
```

```python
## Helper functions
# Assuming 'features' list is defined globally:
# features = ["education-num", "capital-gain", "age", "hours-per-week"]

def eta(x):
    # 1. Move PyTorch batch to CPU and NumPy
    x_np = x.detach().cpu().numpy()

    # # 2. Convert to DataFrame with original feature names
    # x_df = pd.DataFrame(x_np, columns=features)

    # 2. Predict using the Scikit-Learn model (returns NumPy array)
    # probs = best_model.predict_proba(x_df)[:, 1]
    probs = best_model.predict_proba(x_np)[:, 1]

    # 3. Convert back to PyTorch Tensor and move to GPU/Device
    return torch.tensor(probs, dtype=torch.float32, device=device)

@torch.no_grad()
def find_best_response_batch(xb, samples, model, beta, alpha):
    '''
    Finds the best strategic move for a batch of agents xb against a set of samples.
    Adaptable for D-dimensional data.

    Args:
        xb: (B, D) Batch of agents
        samples: (N, D) Database of potential moves
        model: maps (N, D) -> (N,) scores
        beta: scalar budget constraint
        alpha: (D,) vector of cost weights per feature
    '''
    device = xb.device
    dtype = xb.dtype

    # 1. Handle Dimensions (Ensure 2D shapes)
    # If input is 1D (B,), make it (B, 1)
    if xb.dim() == 1:
        xb = xb.unsqueeze(dim=1)
    if samples.dim() == 1:
        samples = samples.unsqueeze(dim=1)

    B, D = xb.shape
    N, D_sample = samples.shape

    # Sanity Check
    assert D == D_sample, f"Dimension Mismatch: Agents have {D} features, Samples have {D_sample}"

    # 2. Process Alpha (Cost Weights)
    # We need alpha to be shape (1, D) to broadcast against (B, N, D)
    alpha_t = torch.as_tensor(alpha, device=device, dtype=dtype)

    if alpha_t.numel() == 1:
        # If scalar alpha given for multi-D, broadcast it to all dims
        alpha_t = alpha_t.view(1, 1).expand(1, D)
    else:
        assert alpha_t.numel() == D, f"Alpha vector must have length {D}, got {alpha_t.numel()}"
        alpha_t = alpha_t.view(1, D)

    beta_t = torch.as_tensor(beta, device=device, dtype=dtype)

    # 3. Calculate Scores for all Samples (Potential Destinations)
    # This gives us the 'utility' of every point in the database
    sample_scores = model(samples).view(-1) # shape: (N,)

    # 4. Calculate Cost Matrix (Vectorized for Multi-D)
    # diff shape: (B, N, D)
    # Broadcasting: (B, 1, D) vs (1, N, D) -> (B, N, D)
    diff = samples.unsqueeze(0) - xb.unsqueeze(1)

    # Cost Function: sum_i( alpha_i * max(0, z_i - x_i) )
    # .clamp(min=0) implements the max(0, ...)
    # .sum(dim=2) aggregates costs across all D features
    cost_matrix = (diff.clamp(min=0) * alpha_t).sum(dim=2) # Result shape: (B, N)

    # 5. Mask Feasible Moves
    is_positive = sample_scores.unsqueeze(0) > 0 # (1, N)
    is_within_budget = cost_matrix <= beta_t     # (B, N)
    feasible_mask = is_positive & is_within_budget

    # 6. Select Best Strategic Move
    # We clone scores to create a masked version for each agent in batch
    masked_scores = sample_scores.unsqueeze(0).expand(B, N).clone()

    # Set infeasible moves to -infinity so they are never chosen
    masked_scores[~feasible_mask] = -float('inf')

    # Find the index of the max score for each agent
    best_indices = masked_scores.argmax(dim=1) # (B,)
    best_strategies = samples[best_indices]    # (B, D)

    # Check if a valid move was actually found (score != -inf)
    max_scores = masked_scores.gather(1, best_indices.unsqueeze(1)).squeeze(1)
    found_feasible = torch.isfinite(max_scores) # (B,)

    # 7. Final Selection
    # If feasible found -> move there. Else -> stay at original position (xb)
    # Unsqueeze found_feasible to (B, 1) to broadcast across D features
    final_strategies = torch.where(found_feasible.unsqueeze(1), best_strategies, xb)

    return final_strategies
```

```python
# --- 1. Device Setup ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Training on: {device}")

# --- 2. Model Setup ---
class LinearClassifier(nn.Module):
    def __init__(self, in_dim, out_dim):
        super().__init__()
        self.out = nn.Linear(in_dim, out_dim, bias=True)

    def forward(self, x):
        return self.out(x)

# --- 3. Trainer Class with Validation ---
class ImprovementAwareStrategicTrainer:
    def __init__(self, model, train_dl, val_dl, opt, scheduler, alpha, beta, samples):
        self.model = model
        self.train_dl = train_dl
        self.val_dl = val_dl
        self.opt = opt
        self.scheduler = scheduler

        # Handle Alpha (Cost Weights) for Multi-D
        # Ensure alpha is a tensor of shape (D,) on the correct device
        if not torch.is_tensor(alpha):
            alpha = torch.tensor(alpha, dtype=torch.float32)
        self.alpha = alpha.to(device)

        self.beta = beta
        self.samples = samples.to(device)

    def oracle(self, xb, yb):
        """
        Label generation process for Multi-D inputs:
        1. Agents move strategically (find_best_response_batch).
        2. Calculate improved Probability.
        3. Sample improved labels.
        """
        xb = xb.to(device)
        yb = yb.to(device).view(-1).long()

        # We use no_grad for the simulation/sampling part
        with torch.no_grad():
            scores = self.model(xb).view(-1)
            # Identify agents who are currently negative (score < 0)
            negative_mask = scores < 0
            xb_strat = xb.clone()

            if negative_mask.any():
                neg_agents = xb[negative_mask]
                # 1. Strategic Movement (Global function assumed)
                # Returns the NEW positions (B_neg, D)
                best_moves = find_best_response_batch(
                    neg_agents, self.samples, self.model, self.beta, self.alpha
                )
                xb_strat[negative_mask] = best_moves

                # 2. Improvement Probability using eta
                eta_curr = eta(xb).view(-1)           # Assuming best_model based on logistic regression is defined globally
                eta_strat = eta(xb_strat).view(-1)

                # Probability of True Label Improvement
                proba_imp = (eta_strat - eta_curr) / (1 - eta_curr + 1e-8)
                proba_imp = proba_imp.clamp(0.0, 1.0)

                # 3. Sample New Labels
                y_imp = yb.clone()
                can_flip_mask = (yb == 0) & negative_mask

                if can_flip_mask.any():
                    flips = torch.bernoulli(proba_imp[can_flip_mask]).long()
                    y_imp[can_flip_mask] = flips

                # Convert to {-1, 1} for margin loss
                y_tilde = (2 * y_imp - 1).float()
            else:
                # If no one moves, y_tilde is just original labels
                y_tilde = (2 * yb - 1).float()
        return xb, y_tilde

    def strat_imp_loss(self, xb, y_tilde):
        """
        Strategic Improvement-Aware Hinge Loss (Multi-D).
        """
        # Model prediction (B,)
        pred = self.model(xb).squeeze(-1)
        w = self.model.out.weight.squeeze() # Shape (D,)

        # Strategic Gain (Scalar)
        # In Multi-D linear model, agent puts budget into feature with max (w_i / alpha_i)
        # We clamp w to 0 because agents won't increase a feature if it has negative weight
        # (Assuming cost is for increasing feature value)
        'can have without torch.relu() the weights,also'
        roi = torch.relu(w) / (self.alpha + 1e-8)
        max_roi = torch.max(roi)
        s_gain = self.beta * max_roi

        # Hinge-like loss with strategic gap
        # If y_tilde is 1 (Positive/Improved), we want Score > 1 - s_gain
        # "Give credit for the potential gain"
        margin = y_tilde * (pred + s_gain)
        loss = torch.relu(1 - margin).mean()
        return loss

    def train(self, n_epochs: int):
        best_val_loss = float("inf")
        history = {'train_loss': [], 'val_loss': []}

        for epoch in tqdm(range(n_epochs)):
            # --- Training Phase ---
            self.model.train()
            train_batch_losses = []

            for xb, yb in self.train_dl:
                self.opt.zero_grad(set_to_none=True)

                # Pass eta_b to oracle
                xb, y_tilde = self.oracle(xb, yb)
                loss = self.strat_imp_loss(xb, y_tilde)

                loss.backward()
                self.opt.step()
                train_batch_losses.append(loss.item())

            avg_train_loss = np.mean(train_batch_losses)
            history['train_loss'].append(avg_train_loss)

            # --- Validation Phase ---
            self.model.eval()
            val_batch_losses = []

            with torch.no_grad():
                for xb, yb in self.val_dl:
                    xb, y_tilde = self.oracle(xb, yb)
                    val_loss = self.strat_imp_loss(xb, y_tilde)
                    val_batch_losses.append(val_loss.item())

            avg_val_loss = np.mean(val_batch_losses)
            history['val_loss'].append(avg_val_loss)

            # --- Scheduler & Checkpointing ---
            self.scheduler.step(avg_val_loss)

            if avg_val_loss < best_val_loss:
                best_val_loss = avg_val_loss
                torch.save(self.model.state_dict(), 'best_model_imp_adult.pth')

            # --- Logging ---
            if epoch % 10 == 0:
                # Multi-D logging: Print norm of weights instead of single threshold
                w_norm = torch.norm(self.model.out.weight).item()
                b_val = self.model.out.bias.item()
                print(f"Ep: {epoch:04d} | Train: {avg_train_loss:.4f} | Val: {avg_val_loss:.4f} | |w|: {w_norm:.4f} | b: {b_val:.4f}")

        return history
```
